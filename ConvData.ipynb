```python
import numpy as np
import cv2 as cv
```

```python
im = cv.imread('Mask_Tank/hoankiem1_tank.png')
imgray = cv.cvtColor(im, cv.COLOR_BGR2GRAY)
ret, thresh = cv.threshold(imgray, 127, 255, 0)
contours, hierarchy = cv.findContours(thresh, cv.RETR_TREE, cv.CHAIN_APPROX_SIMPLE)
```

```python
cnt = contours[0]
x,y,w,h = cv.boundingRect(cnt)
```

```python
from pandas import DataFrame
df = DataFrame(columns = ['image_names', 'class', 'xmin', 'xmax', 'ymin', 'ymax'])
```

```python
for i in range(len(contours)):
    cnt = contours[i]
    x,y,w,h = cv.boundingRect(cnt)
    df = df.append({'image_names': "hoankiem8", 'class': 'tank', 'xmin' : x, 'xmax' : x + w, 'ymin' : y,'ymax': y + h}, ignore_index=True)
```

```python
for i in range(145005):
    im = cv.imread('cropped_mask_tank/hoankiem_{}_tank.png'.format(i))
    imgray = cv.cvtColor(im, cv.COLOR_BGR2GRAY)
    ret, thresh = cv.threshold(imgray, 127, 255, 0)
    contours, hierarchy = cv.findContours(thresh, cv.RETR_TREE, cv.CHAIN_APPROX_SIMPLE)
    for j in range(len(contours)):
        cnt = contours[j]
        x,y,w,h = cv.boundingRect(cnt)
        df = df.append({'image_names': "hoankiem_{}.jpg".format(i), 'class': 'tank', 'xmin' : x, 'xmax' : x + w, 'ymin' : y,'ymax': y + h}, ignore_index=True)
    
```
